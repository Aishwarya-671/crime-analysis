Accoding to the above link 'Part 1-2' is described as Classification of the incident as Part 1 or Part 2 crime or offense.

**Part 1 Offenses:**

 The FBI chose to classify serious crimes when it created a system of Uniform Criminal Reports on Arrests and Offenses, so that we can see the effects for society and public safety. As opposed to so called victimless crimes, these Part I crimes are common throughout the United States and are likely to be reported to the police. By comparing crimes that are known to the police with a total number of arrests, officials can optimise their data on Part I offences. This category includes:

*   Criminal Homicide
*   Rape
*   Robbery
*   Aggravated Assault
*   Burglary
*   Larceny-theft (except motor vehicle theft)
*   Motor Vehicle Theft
*   Arson
*   Human Trafficking, Commercial Sex Acts
*   Human Trafficking, Involuntary Servitude


**Part 2 Offenses:**

Part 2 offenses encompass all other reportable classifications outside those defined as Part I.
In terms of the effect on society and public security, these offences are not as serious as they appear. Due to the fact that victims of Part II crimes are not able to inform police, information on offences known to be committed cannot be used for data purposes; this is also due to the fact that a crime is deemed victimless. Therefore, only offences of Part II are reported for arrest. It includes:

* Other Assaults
* Forgery and Counterfeiting
* Fraud
* Embezzlement
* Stolen Property: Buying, Receiving, Possessing
* Vandalism
* Weapons: Carrying, Possessing, etc.

* Driving Under the Influence
* Liquor Laws
* Drunkenness
* Disorderly Conduct
* Vagrancy
* All Other Offenses

* Curfew and Loitering Laws—(Persons under 18)

**Investigating each column**

Date Rptd: Date when the incident was reported.

```python
# Extract the date
crime_df['Date Rptd'] = pd.to_datetime(crime_df['Date Rptd']).dt.date

# Rename to 'Date Reported'
crime_df = crime_df.rename(columns={'Date Rptd': 'Date Reported'})

# Convert the column to datetime
crime_df['Date Reported'] = pd.to_datetime(crime_df['Date Reported'])
```

We are filtering the crime data in the dataset to focus on a specific time range, starting from January 1, 2020, and ending on September 30, 2023. The purpose of this filtering is to exclude the current month, which is October 2023, in order to conduct an analysis of crime trends up to the present month. This allows us to gather insights into how crime patterns have evolved over the years leading up to September 2023 without being influenced by incomplete data for the current month. By narrowing down the time range, we can obtain a clearer picture of crime trends and make more accurate assessments based on the available historical data.

```python
from datetime import datetime, timedelta

# Define the start date as '2020-01-01'
start_date = '2020-01-01'

# Get the current date
current_date = datetime.now()

# Check if the current date is in or after October 2023
if current_date >= datetime(2023, 10, 1):
    # If it's October 2023 or later, set the end date to the end of September 2023
    end_date = '2023-09-30'
else:
    # Otherwise, set the end date to the current date minus one day
    end_date = (current_date - timedelta(days=1)).strftime('%Y-%m-%d')

# Filter the DataFrame for dates in the specified range
crime_df = crime_df[
    (crime_df['Date Reported'] >= start_date) & (crime_df['Date Reported'] <= end_date)
]
```

```python
# Find the first and last dates in the entire dataset
first_date = crime_df['Date Reported'].min()
last_date = crime_df['Date Reported'].max()

print("First date in the dataset:", first_date)
print("Last date in the dataset:", last_date)
```

The statistics indicate negative and 120 'Victim Age' values, which are not typically meaningful or realistic for age values. It's important to address these values as well. Negative, zero and over 100 ages might be erroneous data entries or missing information or outliers.

```python
# Filter rows with 'Vict Age' less than or equal to 0
crime_df[crime_df['Vict Age'] < 0].shape[0]
```

```python
crime_df[crime_df['Vict Age'] <= 0]
```

```python
crime_df[crime_df['Vict Age'] >100].shape[0]
```

```python
# Calculate the mean of ages above 0
crime_df[crime_df['Vict Age'] > 0]['Vict Age'].mean()
```

```python
# Define a range of values near the mean
age_values = [40, 35, 45, 30]

import numpy as np

# Generate random indices for filling values
random_index = np.random.choice(len(age_values), size=sum(crime_df['Vict Age'] <= 0))

# Impute the values with a random selection of our fixed values
crime_df.loc[crime_df['Vict Age'] <= 0, 'Vict Age'] = np.array(age_values)[random_index]
```

```python
# Remove potential outliers (e.g., one record with 120 years old and missing information)
crime_df = crime_df[crime_df['Vict Age'] <= 100]

# Rename this column to 'Victim Age'
crime_df = crime_df.rename(columns={'Vict Age': 'Victim Age'})

# Basic statistics about the 'Victim Age' column to see what changed
crime_df['Victim Age'].describe()
```

We have taken the following actions to address the above issue:

1. We identified the rows with 'Victim Age' less than or equal to 0.

2. We calculated the mean age of non-zero and non-negative age values.

3. We generated a list of fixed age values around this mean (e.g., 40, 35, 45, 30).

4. Randomly selected values from the list of fixed ages to impute into the rows with negative or zero 'Victim Age.'

5. We also removed one record with an extremely high age of 120, which may be an outlier or erroneous data.

These steps were taken to ensure the 'Victim Age' values are meaningful and more representative of the actual ages of the individuals involved in the reported incidents.

```python
# Create a new derived column that segments ages into groups
age_groups = [
    (crime_df['Victim Age'] <= 12),
    (crime_df['Victim Age'] >= 13) & (crime_df['Victim Age'] < 18),
    (crime_df['Victim Age'] >= 18) & (crime_df['Victim Age'] < 65),
    (crime_df['Victim Age'] >= 65)
]

labels = ['Child', 'Teen', 'Adult', 'Old']

# Create a new column 'Age Group' based on the age segmentation
crime_df['Age Group'] = np.select(age_groups, labels)

# Change the position of the new column to be the 10th column
move = crime_df.pop('Age Group')
crime_df.insert(9, 'Age Group', move)
```

 The purpose of segmenting ages into groups in this code is to categorize individuals based on their age into meaningful and easily interpretable categories. This is likely to have a beneficial effect in the analysis and disclosure of data, as well as providing information about how specific age groups are impacted by or involved in reported crimes. The following groups are to be used in this code:

1. Child: Age 12 or below
2. Teen: Age between 13 and 17
3. Adult: Age between 18 and 64
4. Old: Age 65 or above

 This segmentation will help to understand the age distribution of victims in reported crime, which can be helpful for analysing and reporting different types of criminal activities with a view to finding trends or patterns within specific age groups' participation in any type of criminality.

###Seasonal Patterns:

```python
# Data Preprocessing
crime_df['Date Reported'] = pd.to_datetime(crime_df['Date Reported'])
crime_df['Year'] = crime_df['Date Reported'].dt.year
crime_df['Month'] = crime_df['Date Reported'].dt.month

# Group data by year and month, calculate the average number of crimes per month
monthly_avg_crimes = crime_df.groupby(['Year', 'Month'])['Date Reported'].count().groupby('Month').mean()

# Print the average number of crimes per month
print("Average Number of Crimes per Month:")
monthly_avg_crimes
```

```python
# Create a line chart to visualize the monthly average number of crimes over the years
plt.figure(figsize=(8, 6))
plt.plot(monthly_avg_crimes.index, monthly_avg_crimes.values, marker='o', linestyle='-', color='royalblue', markersize=8, linewidth=2)
plt.title('Monthly Average Number of Crimes (2020-2023)', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Average Number of Crimes', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()

# Display the line chart
plt.show()
```

**Observations:**

> Seasonal trends: data show fluctuations in crime rates over the course of a month. In general, compared to the beginning and the end of the year, there is a pattern of higher crime rates in the middle of the year.

> Summer peak: As warmer weather often leads to increased criminal activity, the months of July and August have the highest average number of offences, which is typical in many regions.

> The winter months are, on the other hand, characterised by a lower average crime rate between November and February. Different factors, including less outdoor activity in the cooler months, are to be taken into account.

> Compared to other winter months, December shows a slight increase in average values which may have been caused by economic activities and gatherings that take place over the holiday period.

> November has the lowest average number of crimes, which may be influenced by the reduced outdoor activities and the approaching holiday season.

The observations indicate that crime rates vary between the warmer and cooler months of the year, suggesting seasonal variations. It may be important to understand these patterns for law enforcement and policy makers in the allocation of resources and planning actions that are effective.

```python
from scipy.stats import chi2_contingency
# Create a subset of the dataset with the chosen crime types
subset_crime_types = [
    'VEHICLE - STOLEN',
    'BATTERY - SIMPLE ASSAULT',
    'THEFT OF IDENTITY',
    'BURGLARY FROM VEHICLE',
    'VANDALISM - FELONY ($400 & OVER, ALL CHURCH VANDALISMS)'
]
subset_df = crime_df[crime_df['Crime Description'].isin(subset_crime_types)]

# Correlation analysis: Chi-squared test for independence

# Age Group vs. Crime Types
age_vs_crime = pd.crosstab(subset_df['Age Group'], subset_df['Crime Description'])
chi2, p, _, _ = chi2_contingency(age_vs_crime)
print('Chi-squared statistic (Age Group vs. Crime Types):', chi2)
print('p-value:', p)

# Gender vs. Crime Types
gender_vs_crime = pd.crosstab(subset_df['Vict Sex'], subset_df['Crime Description'])
chi2, p, _, _ = chi2_contingency(gender_vs_crime)
print('Chi-squared statistic (Gender vs. Crime Types):', chi2)
print('p-value:', p)

# Descent vs. Crime Types
descent_vs_crime = pd.crosstab(subset_df['Victim Descent'], subset_df['Crime Description'])
chi2, p, _, _ = chi2_contingency(descent_vs_crime)
print('Chi-squared statistic (Descent vs. Crime Types):', chi2)
print('p-value:', p)
```

The analysis reveals significant associations between demographic factors (Age Group, Gender, Descent) and specific types of crimes. The chi-squared statistics and p-values suggest that these associations are not occurring by chance. For example, different age groups, genders, and descents exhibit varying patterns of criminal involvement. These findings can be valuable for law enforcement agencies and policymakers to better understand and address the dynamics of criminal activities within distinct demographic groups, allowing for more targeted crime prevention and intervention strategies.

###Predictive Analysis:

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from scipy.stats import norm

# Set 'Date Reported' as the index and resample the data to a specific frequency (e.g., monthly)
crime_data = crime_df.set_index('Date Reported')['Crime Code'].resample('M').sum()

# Print the initial crime data and difference data
print("Initial Crime Data:")
print(crime_data)

# Plot the time series data to visualize it
crime_data.plot(figsize=(12, 6))
plt.title('Monthly Crime Counts')
plt.xlabel('Date Reported')
plt.ylabel('Crime Counts')
plt.show()

# Check for stationarity (Augmented Dickey-Fuller Test or visual inspection)

# Decompose the time series to visualize trends, seasonality, and residuals
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(crime_data, model='additive')
decomposition.plot()
plt.show()

# Determine the order of differencing using ACF and PACF plots
plot_acf(crime_data)
plot_pacf(crime_data)
plt.show()

# Assuming the data is non-stationary, perform differencing to make it stationary
differenced_data = crime_data.diff().dropna()

# Print the differenced data
print("\nDifferenced Data:")
print(differenced_data)

# Check for stationarity again
# You may need to perform differencing multiple times until the data becomes stationary

# Fit the ARIMA model
# Choose the appropriate p, d, and q values based on the ACF and PACF plots
p, d, q = 1, 1, 1  # Adjust these values based on your analysis
model = ARIMA(crime_data, order=(p, d, q))
results = model.fit()

# Forecast future crime trends
forecast_steps = 12  # Adjust the number of steps into the future you want to forecast
forecast = results.forecast(steps=forecast_steps)

# Print forecast values
print("\nForecasted values:")
print(forecast)

# Calculate the standard errors of the forecast
stderr = np.std(results.resid)

# Calculate the confidence intervals
alpha = 0.95  # Adjust as needed
z_score = np.abs(norm.ppf(1 - (1 - alpha) / 2))
conf_int = pd.DataFrame({'lower': forecast - z_score * stderr, 'upper': forecast + z_score * stderr})

# Print standard errors and confidence intervals
print("\nStandard Errors:", stderr)
print("\nConfidence Intervals:")
print(conf_int)

# Plot the observed data and the forecasted data
plt.figure(figsize=(12, 6))
plt.plot(crime_data, label='Observed')
plt.plot(pd.date_range(start=crime_data.index[-1], periods=forecast_steps, freq='M'), forecast, label='Forecast', color='red')
plt.fill_between(pd.date_range(start=crime_data.index[-1], periods=forecast_steps, freq='M'),
                 forecast - stderr, forecast + stderr, color='pink', alpha=0.3, label='Confidence Interval')
plt.legend()
plt.title('Crime Counts Forecast')
plt.xlabel('Date')
plt.ylabel('Crime')
plt.show()
```

Monthly crime count time series data are used in this research. The first step involves visualizing the time series and resampling the data to a monthly frequency. To identify patterns, seasonality, and residuals, the data is broken down. Applying differencing makes the data steady. Future patterns in criminal activity are predicted by fitting an ARIMA model. Calculating standard errors and confidence intervals helps evaluate how reliable the forecast is.
